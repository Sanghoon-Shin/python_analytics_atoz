- 서울시 대기 오염 공개 정보를 이용하여 데이터 분석을 위한 Pandas의 기능을 알아 보았다

- visualization : matplotlib, graph 그리는 template, 한글 표시 팁, seaborn, bokeh, plotly, etc.
